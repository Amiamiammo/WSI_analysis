```python
import pickle
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
pickle_file_resnet50 = "embeddings_patch_library/resnet50.pkl"
```

```python
# Load the data from the pickle file
with open(pickle_file_resnet50, 'rb') as handle:
    data = pickle.load(handle)

# If the data is not already a numpy array, convert it
if not isinstance(data, np.ndarray):
    embeddings = np.array(data['embeddings'])
    labels = np.array(data['labels'])
```

```python
# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(embeddings)
```

```python
# Apply PCA
pca = PCA()
pca.fit(data_scaled)

# Calculate cumulative variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Determine the number of components that explain 99% of the variance
components_99_variance = np.argmax(cumulative_variance >= 0.99) + 1
```

```python
# Plot cumulative variance
plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance, marker='o', linestyle='--', color='b')

# Add horizontal line at 99% variance
plt.axhline(y=0.99, color='r', linestyle='--', label='99% variance')

# Add vertical line at the number of components for 99% variance
plt.axvline(x=components_99_variance, color='g', linestyle='--', label=f'{components_99_variance} components')

# Annotations
plt.title('Cumulative Variance Explained by PCA Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.grid(True)
plt.legend()
plt.show()
```
